### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
import re 

def cleanup(text):

    #import string
    punctuation = '[!-.:?;/@]' # Note que os sinais [] s√£o delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    return text_subbed
```

```python
train.loc[:,'Classifica√ß√£o'] = train['Classifica√ß√£o'].astype('category')
train['Classifica√ß√£o'].cat.categories = ["Irrelevante", "Relevante"]
train
```

```python
#relevantes:
relevantes_ = train["Classifica√ß√£o"] == "Relevante"
relevantes = train.loc[relevantes_,["Treinamento"]].copy()
relevantes
```

```python
#irrelevantes:
irrelevantes_ = train["Classifica√ß√£o"] == "Irrelevante"
irrelevantes = train.loc[irrelevantes_,["Treinamento"]].copy()
irrelevantes
```

```python
lista_rel = relevantes["Treinamento"].tolist()
lista_rel
```

```python
lista_irrel = irrelevantes["Treinamento"].tolist()
lista_irrel
```

```python
texto_rel = ''.join(lista_rel)
texto_rel
```

```python
texto_irrel = ''.join(lista_irrel)
texto_irrel
```

```python
rel_clean = cleanup(texto_rel)
rel_clean
```

```python
irrel_clean = cleanup(texto_irrel)
irrel_clean
```

```python
texto_total = rel_clean + irrel_clean
texto_total
```

```python
texto_rel_s = rel_clean.split()
texto_rel_s
```

```python
texto_irrel_s = irrel_clean.split()
texto_irrel_s
```

```python
texto_total_s = texto_total.split()
texto_total_s
```

```python
dataframe_rel = pd.Series(texto_rel_s)
dataframe_rel
```

```python
dataframe_irrel = pd.Series(texto_irrel_s)
dataframe_irrel
```

```python
dataframe_total = pd.Series(texto_total_s)
dataframe_total
```

```python
freq_abs_rel = dataframe_rel.value_counts().round(4)
freq_abs_rel
```

```python
freq_abs_irrel = dataframe_irrel.value_counts().round(4)
freq_abs_irrel
```

```python
freq_abs_total = dataframe_total.value_counts().round(4)
freq_abs_total
```

```python
#Fun√ß√£o Naive Bayes
lista_prob_rel = []
for i in test['Teste']:
    P_R = 1
    i = i.split()
    for j in i:
        if j in freq_abs_rel:
            c = freq_abs_rel[j]        
        else:
            c = 0
        probabilidade = (c + 1)/(len(dataframe_rel) + len(freq_abs_total))
        P_R *= probabilidade
    lista_prob_rel.append(P_R)
lista_prob_rel
```

```python
lista_prob_irrel = []
for i in test['Teste']:
    P_IR = 1
    i = i.split()
    for j in i:
        if j in freq_abs_irrel:
            c = freq_abs_irrel[j]        
        else:
            c = 0
        prob = (c + 1)/(len(dataframe_irrel) + len(freq_abs_total))
        P_IR *= prob
    lista_prob_irrel.append(P_IR)
lista_prob_irrel
```

```python
Naive_Bayes = []
for i in range(0,len(lista_prob_rel)):
    if (lista_prob_rel[i] > lista_prob_irrel[i]):
        Naive_Bayes.append('Relevante') 
    else:
        Naive_Bayes.append('Irrelevante') 
        
test["Naive Bayes"]= Naive_Bayes
test.head(40)
```

```python
tabela = pd.crosstab(test["Classifica√ß√£o"], test["Naive Bayes"], normalize = "columns")*100
tabela
```

Falsos positivos = 38.6%

Falsos negativo = 54.3%
