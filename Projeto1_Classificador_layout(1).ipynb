### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
import re 

def cleanup(text):

    #import string
    punctuation = '[!-.:?;/@]' # Note que os sinais [] s√£o delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    return text_subbed
```

```python
train.loc[:,'Classifica√ß√£o'] = train['Classifica√ß√£o'].astype('category')
train['Classifica√ß√£o'].cat.categories = ["Irrelevante", "Relevante"]
train
```

```python
#relevantes:
relevantes_ = train["Classifica√ß√£o"] == "Relevante"
relevantes = train.loc[relevantes_,["Treinamento"]].copy()
relevantes
```

```python
#irrelevantes:
irrelevantes_ = train["Classifica√ß√£o"] == "Irrelevante"
irrelevantes = train.loc[irrelevantes_,["Treinamento"]].copy()
irrelevantes
```

```python
lista_rel = relevantes["Treinamento"].tolist()
lista_rel
```

```python
lista_irrel = irrelevantes["Treinamento"].tolist()
lista_irrel
```

```python
texto_rel = ''.join(lista_rel)
texto_rel
```

```python
texto_irrel = ''.join(lista_irrel)
texto_irrel
```

```python
rel_clean = cleanup(texto_rel)
rel_clean
```

```python
irrel_clean = cleanup(texto_irrel)
irrel_clean
```

```python
texto_total = rel_clean + irrel_clean
texto_total
```

```python
texto_rel_s = rel_clean.split()
texto_rel_s
```

```python
texto_irrel_s = irrel_clean.split()
texto_irrel_s
```

```python
texto_total_s = texto_total.split()
texto_total_s
```

```python
dataframe_rel = pd.Series(texto_rel_s)
dataframe_rel
```

```python
dataframe_irrel = pd.Series(texto_irrel_s)
dataframe_irrel
```

```python
dataframe_total = pd.Series(texto_total)
dataframe_total
```

```python
freq_abs_rel = dataframe_rel.value_counts().round(4)
freq_abs_rel
```

```python
freq_abs_irrel = dataframe_irrel.value_counts().round(4)
freq_abs_irrel
```

```python
freq_abs_total = dataframe_total.value_counts().round(4)
freq_abs_total
```
